```python
'''prepare dataset'''
df=pd.read_csv('adult_dataset.csv', delimiter=',') ## original dataset
df.shape
```

```python
df.columns
```

```python
df=df[(df.age>20) & (df.age<60)]
df=df[['id','age','education','marital status','gender','salary','race']]
df.to_csv('dataset.csv', index=False)
df.shape
```

```python
def generalize(df):
    df['age'] = df.apply(lambda row: '[{}-{}]'.format(int(row.age/10)*10,
                                                            (int(row.age/10)+1)*10), axis=1)
    df['gender'] = '--'
    df['education'] = df.apply(lambda row: 
                    'Grad' if row.education in [' Masters', ' Bachelors'] else 'Ugrad', axis=1)
    df['marriage'] = df.apply(lambda row: 
                    'Never-married' if row['marital status']== ' Never-married' else 'Once-married', axis=1)
    df['race'] = df.apply(lambda row: 
                    'Race1' if row['race']== ' White' else 'Race2', axis=1)
    return df
dfg = generalize(df)
dfg
```

```python
dfg=dfg[['id','age','education','marriage','gender','salary','race']].reset_index(drop=True)
dfg.to_csv('anon-dataset.csv', index=False)
```
